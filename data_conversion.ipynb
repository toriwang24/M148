### Iterate through directory

```python
n = 0 # Keep track of the number of files being converted.
df_dict = {}

for file_name in os.listdir(file_path):
    f = os.path.join(file_path, file_name)
    if file_name[0] != '.': # Ignores '.DS_Store' files
        key = 'df_' + str(file_name)
        df_dict[key] = pd.read_parquet(f)
        n = n + 1
        print('Converted ' + str(n) + ' files')
```
