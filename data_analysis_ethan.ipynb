# Statistics of product list for checkout 

```python
%%time
#Collect columns from all parquet files into a pandas DataFrame

columns = ['checkoutthankyouflag', 'sitesection', 'productlist']
dfs = []
for member in tar:
    if member.isreg():
        df_temp = pd.read_parquet(member.name,columns = columns)
        dfs.append(df_temp)

df_all = pd.concat(dfs)
```

# More basic stats

```python
%%time
#Collect columns from all parquet files into a pandas DataFrame

columns = ['evar23', 'geocity', 'geocountry', 'georegion']
dfs = []
for member in tar:
    if member.isreg():
        df_temp = pd.read_parquet(member.name,columns = columns)
        dfs.append(df_temp)

df_all = pd.concat(dfs)
```

```python
df_subset = df_all.drop_duplicates(subset=['evar23'])
df_subset = df_subset.iloc[1: , :]
df_subset_usa = df_subset.loc[df_subset['geocountry'] == 'usa']
```

```python
proportion_city = df_subset_usa.value_counts('geocity').rename_axis('city').reset_index(name='proportion')
df_city = pd.DataFrame(proportion_city)
df_city['proportion'] = df_city['proportion'] / len(df_subset_usa.index)
df_city.head(10)
```

```python
df_subset_usa.georegion.unique().sort()
proportion = df_subset_usa.value_counts('georegion').rename_axis('state_code').reset_index(name='proportion')
df_heatmap = pd.DataFrame(proportion)
df_heatmap['proportion'] = df_heatmap['proportion'] / len(df_subset_usa.index)
df_heatmap['state_code'] = df_heatmap['state_code'].str.upper()
df_heatmap.head(10)
```

```python
import plotly.express as px
fig = px.choropleth(df_heatmap,
                    locations='state_code', 
                    locationmode="USA-states", 
                    scope="usa",
                    color = 'proportion',
                    color_continuous_scale="Viridis_r"
                    )
fig.show()
```
